# 2. Basic Analysis

Question 2.1: What is the total number of orders, total sales, and total profit in the dataset?

Question 2.2: Which customer segment (Segment) has the highest number of orders? Visualize this using a bar chart.

```python
segment_orders = super_store_df.groupby('Segment')['Order ID'].nunique().sort_values(ascending = False)
print(segment_orders , 'segment_orders')
```

```python
# Plot number of orders by segment
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
plt.bar(segment_orders.index, segment_orders.values,color = colors)
plt.xlabel('Segment')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Segmentnt')
plt.show()
```

```python
region_orders = super_store_df.groupby('Region')['Order ID'].nunique().sort_values(ascending = False)
print(region_orders , 'region_orders')
```

```python
colors = ['red','green','blue','purple']
plt.bar(region_orders.index,region_orders.values,color = colors)
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Region')
plt.show()
```

# 3. Time-Based Analysis

```python
super_store_df['Year'] = super_store_df['Order Date'].dt.year
sum_of_sales_per_year = super_store_df.groupby('Year')['Sales'].sum()
sum_of_sales_per_year = pd.DataFrame(sum_of_sales_per_year)
sum_of_sales_per_year
```

```python

sum_of_profit_per_year = super_store_df.groupby('Year')['Profit'].sum()
sum_of_profit_per_year = pd.DataFrame(sum_of_profit_per_year)
sum_of_profit_per_year
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot total sales over the years
plt.plot(sum_of_sales_per_year.index, sum_of_sales_per_year['Sales'], marker='o', label='Total Sales', color='blue')

# Plot total profit over the years
plt.plot(sum_of_profit_per_year.index, sum_of_profit_per_year['Profit'], marker='o', label='Total Profit', color='brown')

# Add titles and labels
plt.title('Total Sales and Profit Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Question 3.2: Which month has the highest number of orders? Create a bar chart to visualize monthly order counts.

```python
# Extract the month names from the 'Order Date' column
super_store_df['Month'] = super_store_df['Order Date'].dt.month_name()

# Group by 'Month' and count the unique 'Order ID', then sort by the month order
monthly_oders = (
    super_store_df.groupby('Month')['Order ID']
    .nunique()
    .sort_index(key=lambda x: pd.to_datetime(x, format='%B').month)
)

# Print the result with a proper new line character
print(monthly_oders, '\n')
```

```python
plt.bar(monthly_oders.index,monthly_oders.values)
plt.xlabel('Month')
plt.xticks(rotation = 45)
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Month')
plt.show()
```

Question 3.3: What is the average shipping time (Ship Date - Order Date) in days? Plot a histogram of shipping times.

```python
shipping_days = (super_store_df['Ship Date'] - super_store_df['Order Date']).dt.days
average_shipping_days  = shipping_days.mean()
average_shipping_days.round(2)
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot the histogram of shipping times
plt.hist(average_shipping_days, bins=30, color='orange', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Shipping Times')
plt.xlabel('Shipping Time (Days)')
plt.ylabel('Frequency')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Product and Category Analysis

Question 4.1: Which product category (Category) has the highest sales and profit? Visualize this using a bar chart.

```python
category_sales = super_store_df.groupby('Category')['Sales'].sum()
category_sales = pd.DataFrame(category_sales).reset_index()
category_sales
```

```python
category_profit = super_store_df.groupby('Category')['Profit'].sum()
category_profit = pd.DataFrame(category_profit).reset_index()
category_profit

category_sales_and_profits = pd.merge(category_sales,category_profit,on = 'Category')
category_sales_and_profits
```

```python
plt.figure(figsize=(10,6))
fig,ax1= plt.subplots()
ax2 = ax1.twinx()
ax1.bar(category_sales_and_profits['Category'],category_sales_and_profits['Profit'],color='green')
ax2.plot(category_sales_and_profits['Category'],category_sales_and_profits['Sales'], color = 'orange')

ax1.set_xlabel('Category')
ax1.set_ylabel('Profit')
ax2.set_ylabel('Sales')
plt.title('Sales and Profit by Category')
plt.legend()
plt.show()
```

Question 4.2: What are the top 10 products by sales? Create a bar chart to display them.

```python
super_store_df.head(5)
```

```python

top_products = super_store_df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products = pd.DataFrame(top_products).reset_index()
top_products
```

Question 4.3: Which sub-category (Sub-Category) has the highest average discount? Visualize this using a bar chart.

```python
sub_category = super_store_df.groupby('Sub-Category')['Discount']
highest_average_discount = sub_category.mean().sort_values(ascending = False)
highest_average_discount = pd.DataFrame(highest_average_discount).reset_index()
highest_average_discount
```

```python
plt.bar(highest_average_discount['Sub-Category'],highest_average_discount['Discount'])
plt.xlabel('Sub-Category')
plt.ylabel('Average Discount')
plt.title('Average Discount by Sub-Category')
plt.xticks(rotation = 90)
plt.show()
```
